```python
from scipy import stats

pval = []
sum_games = []

for i in range(1,300):
    
    games = i
    wins = games * 0.59
    sum_games.append(games)
    pval.append(stats.binom_test(int(wins), n=games, p=0.4993, alternative = "two-sided"))

# Es gibt eine 27,92% Chance, dass Megantos 24 aus 41 Games gewinnt, wenn die Chance zu gewinnen bei 49,93% liegt.
# Wenn die Chance(27,92) unter 5% liegen würde, also unter dem Threshold, dann würden wir annehmen,
#dass die Gewinnchance nicht bei 49,93% liegt 
#und die allgemeine Winrate und Megantos Winrate signifikant unterschiedlich sind.-> ergo er hätte Impact
```

```python
from matplotlib import pyplot as plt

plt.plot(sum_games, pval)
plt.axvline(41, color="r", label="aktuelle samplesize 41 games")
plt.axvline(129, color="g", label="samplesize für impact Bed: gleiche WR")
plt.legend()
plt.xlabel("Anzahl Games")
plt.ylabel("pval")
plt.show()
```
